```python
path = './images/images/'
train_path = './images/images/train'
valid_path = './images/images/valid'
angry_file = './images/images/valid/angry/0'
```

```python
# vgg16 = models.vgg16_bn(pretrained=True)
```

```python
learn=cnn_learner(data,models.vgg16_bn,metrics=[error_rate, accuracy]).to_fp16()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4,max_lr=slice(1e-3,1e-1), callbacks=[CSVLogger(learn), ShowGraph(learn)])
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
save_callback = SaveModelCallback(learn, every='epochSaveModelCallbackonitor='accuracy')
```

```python
learn.fit_one_cycle(6,max_lr=slice(1e-6,1e-4), callbacks=[CSVLogger(learn), save_callback, ShowGraph(learn)])
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
def plot_graphs(model):
    model.recorder.plot_losses()
    model.recorder.plot_lr(show_moms=True)
    model.recorder.plot_metrics()
```

```python
plot_graphs(learn)
```

```python
learn.fit_one_cycle(6,max_lr=slice(1e-6,1e-4), callbacks=[CSVLogger(learn), save_callback, ShowGraph(learn)])
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(6,max_lr=slice(1e-7,1e-6), callbacks=[save_callback, ShowGraph(learn)])
```
